## Describe new parameters in the inspections


In particular we rely on the function ```resource.getrusage(who)```. This function returns an object that describes the resources consumed by either the current process or its children, as specified by the who parameter. 

The fields ru_utime and ru_stime of the return value are floating point values representing the amount of time spent executing in user mode and the amount of time spent executing in system mode, respectively. The remaining values are integers. 
A summary of the parameters is presented below:

```
===============================================================================
===============================================================================

ru_utime    | time in user mode (float)

       ru_utime
              This is the total amount of time spent executing in user mode,
              expressed in a timeval structure (seconds plus microseconds). 
              In other words, user time is CPU time spent executing the user 
              program, rather than in kernel system calls.

===============================================================================
===============================================================================

ru_stime    | time in system mode (float)

       ru_stime
              This is the total amount of time spent executing in kernel
              mode, expressed in a timeval structure (seconds plus microsec‐
              onds). In other words, system time is the CPU time spent executing 
              system calls on behalf of the process.

===============================================================================
===============================================================================

ru_maxrss   | maximum resident set size (int)

       ru_maxrss (since Linux 2.6.32)
              This is the maximum resident set size used (in kilobytes).
              For RUSAGE_CHILDREN, this is the resident set size of the
              largest child, not the maximum resident set size of the
              process tree.

NOTE: A process's resident set size is the amount of memory that belongs to it and is currently present (resident) in RAM (real RAM, not swapped or otherwise not-resident).

===============================================================================

ru_ixrss    | shared memory size (int)

       ru_ixrss (unmaintained)
              This field is currently unused on Linux.

===============================================================================

ru_idrss    | unshared memory size (int)

       ru_idrss (unmaintained)
              This field is currently unused on Linux.

===============================================================================

ru_isrss    | unshared stack size (int)

       ru_isrss (unmaintained)
              This field is currently unused on Linux.

===============================================================================

ru_minflt   | page faults not requiring I/O (int)

       ru_minflt
              The number of page faults serviced without any I/O activity;
              here I/O activity is avoided by “reclaiming” a page frame from
              the list of pages awaiting reallocation.

===============================================================================

ru_majflt   | page faults requiring I/O (int)

       ru_majflt
              The number of page faults serviced that required I/O activity.

===============================================================================

ru_nswap    | number of swap outs (int)

       ru_nswap (unmaintained)
              This field is currently unused on Linux.

===============================================================================

ru_inblock  | block input operations (int)

       ru_inblock (since Linux 2.6.22)
              The number of times the filesystem had to perform input.

===============================================================================

ru_oublock  | block output operations (int)

       ru_oublock (since Linux 2.6.22)
              The number of times the filesystem had to perform output.

===============================================================================

ru_msgsnd   | messages sent (int)

       ru_msgsnd (unmaintained)
              This field is currently unused on Linux.

===============================================================================

ru_msgrcv   | messages received (int)

       ru_msgrcv (unmaintained)
              This field is currently unused on Linux.

===============================================================================
ru_nsignals | signals received (int)

       ru_nsignals (unmaintained)
              This field is currently unused on Linux.

===============================================================================
===============================================================================

ru_nvcsw    | voluntary context switches (int)

       ru_nvcsw (since Linux 2.6)
              The number of times a context switch resulted due to a process
              voluntarily giving up the processor before its time slice was
              completed (usually to await availability of a resource).

===============================================================================
===============================================================================

ru_nivcsw   | involuntary context switches (int)

       ru_nivcsw (since Linux 2.6)
              The number of times a context switch resulted due to a higher
              priority process becoming runnable or because the current
              process exceeded its time slice.
              
===============================================================================
===============================================================================
```

with other systems, and because they may one day be supported on

````python
def filter_dfs(df_s, filter_df):
    """Filter the specific dataframe created for a certain key, combination of keys or for a tree depth"""
    if type(filter_df) is str:
        available_keys = set(df_s["Current_key"].values)
        available_combined_keys = set(df_s["Upper_keys"].values)

        if filter_df in available_keys:
            ndf = df_s[df_s["Current_key"].str.contains(f"^{filter_df}$", regex=True)]

        elif filter_df in available_combined_keys:
            ndf = df_s[df_s["Upper_keys"].str.contains(f"{filter_df}$", regex=True)]
        else:
            print("The key is not in the json")
            ndf = "".join(
                [
                    f"The available keys are (WARNING: Some of the keys have no leafs):{available_keys} ",
                    f"The available combined keys are: {available_combined_keys}",
                ]
            )
    elif type(filter_df) is int:
        max_depth = df_s["Tree_depth"].max()
        if filter_df <= max_depth:
            ndf = df_s[df_s["Tree_depth"] == filter_df]
        else:
            ndf = f"The maximum tree depth available is: {max_depth}"
    return ndf
````

````python
def create_histogram(data: pd.DataFrame,
                     columns: Union[str, List[str]] = None,
                     title_hist: str = "Histogram plot",
                     x_label: str = "Variable [Measurement Unit]",
                     bins: int = None, **kwargs):
    """Create Histogram plot."""
    columns = columns if columns is not None else data[columns].columns

    if not bins:
        bins = np.max([np.lib.histograms._hist_bin_auto(data[col].values, None) for col in columns])

    figure = data[columns].iplot(
        title=title_hist,
        yTitle="count",
        xTitle=x_label,
        kind="hist",
        bins=int(np.ceil(bins)),
        asFigure=True,
    )

    return figure
````

````python
def create_box(data: pd.DataFrame,
               columns: Union[str, List[str]] = None,
               title_box: str = "Box plot",
               y_label: str = "Variable [Measurement Unit]", **kwargs):
    """Create duration Box plot."""
    columns = columns if columns is not None else data[columns].columns

    figure = data[columns].iplot(
        kind="box", title=title_box, yTitle=y_label, asFigure=True
    )

    return figure
````

````python
def create_scatter_and_correlation(data: pd.DataFrame,
                                  columns: Union[str, List[str]] = None,
                                  title_scatter: str = "Scatter plot"):
    """Create Scatter plot and evaluate correlation coefficients."""
    columns = columns if columns is not None else data[columns].columns

    figure = data[columns].iplot(
        kind="scatter",
        x=columns[0],
        y=columns[1],
        title=title_scatter,
        xTitle=columns[0],
        yTitle=columns[1],
        mode='markers',
        asFigure=True
    )

    corr_pearson = data[columns].corr("pearson")
    print()
    print("Pearson correlation results:")
    print(corr_pearson)

    corr_spearman = data[columns].corr("spearman")
    print()
    print("Spearman correlation results:")
    print(corr_spearman)
    
    corr_kendall = data[columns].corr("kendall")
    print()
    print("Kendall correlation results:")
    print(corr_kendall)
    
    return figure
````

````python
def color_negative_red(val):
    """
    Takes a scalar and returns a string with
    the css property `'color: red'` for negative
    strings, black otherwise.
    """
    color = 'red' if  val < -0.88 or val > 0.88 else 'black'
    return 'color: %s' % color
````

For this analysis we are going to focus on the following parameters obtained from usage:

- `ru_utime`
- `ru_stime`

**Background**

A system has two "privilege levels":
- `kernel mode`
- `user mode`

Basically a program is not allowed to touch the hardware (like the disk) directly; it has to ask the operating system to do it. 
Therefore, user mode cannot see all of memory, it cannot communicate with I/O devices, etc.. 
For anything more complicated (e.g. memory page allocation, filesystem read/write, printing things to the screen), it must ask the kernel to do it instead, which has access to all of this. 
This is done through a mechanism called "system calls".

Some of the time that a CPU is working is spent running your program, and some of the time is used doing things on your behalf (like reading from the disk, or from the keyboard).
The time running your program is tracked under "user time" and the time spent doing things on your behalf is "system time". This is the `utime` and `stime`.
At a high level, the `ru_utime` struct returns the amount of time your program has spent doing actual computation, and the `ru_stime` struct returns the amount of time your program has been waiting for an answer from the kernel when doing disk access, printing to the screen etc.
